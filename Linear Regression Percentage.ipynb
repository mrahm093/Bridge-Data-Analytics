```python
# Imports
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import seaborn as sns
import sklearn
import numpy as np
```

```python
# Import data
env_nbe_nbi_df = pd.read_excel('Base Model_2019_ML_Ashiqur.xlsx', sheet_name='Base Model')
print(env_nbe_nbi_df.head())
```

```python
env_nbe_nbi_df = env_nbe_nbi_df[env_nbe_nbi_df['Steel_Element'] == 107] 
print(env_nbe_nbi_df)
```

# Normalized

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
scale_features = ['Bridge_Number', 'Steel_Element', 'CS1', 'CS2', 'CS3', 'CS4',
       'temp_dp_2m', 'avg_rh_2m', 'temp_a_10m', 'temp_a_2m', 'temp_a_60c',
       'temp_s_2cm', 'ppt_2019', 'hplus_2019', 'SO4_2019', 'Cl_2019',
       'totalN_201', 'NO3_2019', 'NH4_2019', 'Na_2019', 'K_2019', 'Mg_2019',
       'Ca_2019', 'MAINTENANCE_021', 'Bridge_Age', 'ADT_029',
       'DEGREES_SKEW_034', 'SERVICE_ON_042A', 'SERVICE_UND_042B',
       'STRUCTURE_KIND_043A', 'STRUCTURE_TYPE_043B', 'APPR_KIND_044A',
       'APPR_TYPE_044B', 'DECK_COND_058', 'SUPERSTRUCTURE_COND_059',
       'SUBSTRUCTURE_COND_060', 'OPERATING_RATING_064', 'INVENTORY_RATING_066',
       'STRUCTURAL_EVAL_067', 'APPR_ROAD_EVAL_072', 'DECK_STRUCTURE_TYPE_107',
       'SURFACE_TYPE_108A', 'MEMBRANE_TYPE_108B', 'DECK_PROTECTION_108C',
       'PERCENT_ADT_TRUCK_109','BRIDGE_CONDITION']
env_nbe_nbi_df[scale_features] = ss.fit_transform(env_nbe_nbi_df[scale_features])
```

```python
env_nbe_nbi_df 
```

# 3D

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
fig.set_size_inches(12.5, 7.5)
ax = fig.add_subplot(111, projection='3d')

para_X = 'SO4_2019'
para_Y = 'Na_2019'
para_Z = 'CS1'

env_nbe_nbi_df = env_nbe_nbi_df[env_nbe_nbi_df['NO3_2019'] <=6000]
ax.scatter(xs=env_nbe_nbi_df[para_X], ys=env_nbe_nbi_df[para_Y], zs=env_nbe_nbi_df[para_Z])

ax.set_ylabel(para_Y); ax.set_xlabel(para_X); ax.set_zlabel(para_Z)
ax.view_init(10, -60)
```

### Feature for prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = env_nbe_nbi_df[[
       'temp_dp_2m', 'avg_rh_2m', 'temp_a_10m', 'temp_a_2m', 'temp_a_60c',
       'temp_s_2cm', 'ppt_2019', 'hplus_2019', 'SO4_2019', 'Cl_2019',
       'totalN_201', 'NO3_2019', 'NH4_2019', 'Na_2019', 'K_2019', 'Mg_2019',
       'Ca_2019', 
#     'MAINTENANCE_021', 
    'Bridge_Age', 'ADT_029',
#        'DEGREES_SKEW_034', 
#     'SERVICE_ON_042A', 
#     'SERVICE_UND_042B',
#        'STRUCTURE_KIND_043A', 
#     'STRUCTURE_TYPE_043B', 
#     'APPR_KIND_044A',
#        'APPR_TYPE_044B', 
#     'DECK_COND_058', 
#     'SUPERSTRUCTURE_COND_059',
#        'SUBSTRUCTURE_COND_060', 
    'OPERATING_RATING_064', 'INVENTORY_RATING_066',
#        'STRUCTURAL_EVAL_067', 
#     'APPR_ROAD_EVAL_072', 
#     'DECK_STRUCTURE_TYPE_107',
#        'SURFACE_TYPE_108A', 'MEMBRANE_TYPE_108B', 'DECK_PROTECTION_108C',
#        'PERCENT_ADT_TRUCK_109',
    'BRIDGE_CONDITION']].values
Y = env_nbe_nbi_df[['CS1','CS2','CS3','CS4']].values

X_train, X_test, y_train, y_test = np.asarray(train_test_split(X, Y, test_size=0.3))
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
#predict
result = reg.predict(X_test)
print('The result of predict \n',result)
print('The result of original \n',y_test)
```

```python
# Score
s1 = reg.score(X_train,y_train)
s2 = reg.score(X_test,y_test)
print('The accuracy in training model',s1)
print('The accuracy in testing',s2)



# The accuracy in training model 0.16342864437701546
# The accuracy in testing 0.08483309328321101
```
